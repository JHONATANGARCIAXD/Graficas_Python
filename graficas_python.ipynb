# GRAFICAS CON LA LIBRERIA MATPLOTLIB

## **GRAFICAS EN DOS DIMENSIONES**


**Gr√°fica de una lista**

```python
# Importamos la libreria 
# import matplotlib.pyplot

from matplotlib.pyplot import plot, show, xlim, ylim, xticks, legend, figure
from numpy import linspace, pi , sin, cos
```

```python
# Grafica No.1

y = [1,2,3,4,5]

plot(y)
show()
```

```python
# Grafica No.2 

y = [1,2,3,4,5]

plot(y, '^')
show()
```

```python
# Grafica No. 3

x= [1,2,3,4,5]
y = [2,4,6,8,10]

plot(x, y, '^')
show()  
```

```python
# Grafica No. 4

# Grafica con limites

x= [1,2,3,4,5]
y = [2,4,6,8,10]

plot(x, y, '^')
xlim(-10,10)
ylim(0, 15)
show()
```

```python
# Grafica No. 5 

x = linspace(-10*pi, 10*pi, 2560)
y = sin(x)

plot(x,y)
ylim(-2,2)
show()
```

```python
# Grafica No. 6 

x = linspace(-pi, pi, 256)
y = sin(x)

plot(x,y)
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```

```python
# Grafica No. 7

x = linspace(-pi, pi, 256)
y = sin(x)
z = cos(x)
plot(x,y, label = 'seno')
plot(x,z, label = 'Coseno')
legend(loc='upper left')
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```

```python
# Grafica No. 8

x = linspace(-pi, pi, 256)
y = sin(x)
z = cos(x)

# Figura 1
a = figure(2)
plot(x,y, label = 'seno')
legend(loc='upper left')
# Figura 2
b = figure(20)
plot(x,z, label = 'Coseno')
legend(loc='upper left')
ylim(-2,2)
xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
show()
```
